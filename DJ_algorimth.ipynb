# Implementação do algoritmo Deutsch-Jozsa

## Importações

```python
import qiskit
from qiskit import QuantumCircuit
```

## Função Balanceada

### Porta CNOT, ou função XOR

```python
qc = QuantumCircuit(5,4)

qc.x(4)

for i in range(5):
    qc.h(i)

qc.barrier()
for i in range(0,4):
    qc.cx(i,4)

qc.barrier()

for i in range(4):
    qc.h(i)
qc.barrier()

qc.measure([0,1,2,3], [0,1,2,3])
```

```python
qc.draw('mpl')
```

```python
from qiskit.primitives import Sampler
from qiskit.visualization import plot_distribution

Sampler = Sampler()
job = Sampler.run(circuits = [qc])
job_result = job.result()

prob = [q.binary_probabilities() for q in job_result.quasi_dists]
prob = {chave:i[chave].item() for i in prob for chave in i}
plot_distribution(prob, figsize = (16,8))
```

P(0000) = 0, portanto, a função é balanceada, como esperado.

## Função Constante

### Operador Identidade

```python
qc = QuantumCircuit(5,4)

qc.x(4)

for i in range(5):
    qc.h(i)

qc.barrier()

for i in range(4):
    qc.h(i)
qc.barrier()

qc.measure([0,1,2,3], [0,1,2,3])

qc.draw('mpl')
```

```python
from qiskit.primitives import Sampler
from qiskit.visualization import plot_distribution

Sampler = Sampler()
job = Sampler.run(circuits = [qc])
job_result = job.result()

prob = [q.binary_probabilities() for q in job_result.quasi_dists]
prob = {chave:i[chave].item() for i in prob for chave in i}
plot_distribution(prob, figsize = (16,8))
```

P(0000) = 1, portanto, a função é constante, como esperado.
